```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 創建基本地圖
# "projection" 參數為“ortho"時，所得圖位地球儀截面 

# 設定圖形大小
plt.figure( figsize=(5,5) )

# 設定投影型態
m = Basemap( projection='ortho', resolution=None, lat_0=0, lon_0=100 )

# 設定圖形顏色，選用 NASA 寶石藍輸出 
m.bluemarble(scale=0.5)

# 秀圖
plt.show()
```

```python
# 創建基本地圖
# "projection" 參數為“Mill"，所得平展位面 

# 設定圖形大小
plt.figure( figsize=(5,5) )

# 設定投影型態
m = Basemap( projection='mill', resolution=None, lat_0=0, lon_0=100 )

# 設定圖形顏色，選用 NASA 寶石藍輸出 
m.bluemarble(scale=0.5)

# 秀圖
plt.show()
```

```python
# 創建基本地圖
# "projection" 參數為"lcc"，並以NASA寶石藍輸出

# 設定圖形大小
plt.figure( figsize=(5,5) )

# 設定投影型態
m = Basemap(width=12000000,height=9000000,projection='lcc',
            resolution=None,lat_1=45.,lat_2=55,lat_0=50,lon_0=100)

# 設定圖形顏色，選用 NASA 寶石藍輸出 
m.bluemarble(scale=0.5)

# 秀圖
plt.show()
```

```python
#  "projection" 參數為“Mill"
# 設定經緯座標
# 左下角的緯度 - llcrnrlat = -90,
# 左下角的經度 - llcrnrlon = -180,
# 右上角的緯度 - urcrnrlat = 90,
# 右上角的經度 -urcrnrlon = 180

# 成圖設定resolution

plt.figure( figsize=(16,12) )
m = Basemap(projection='mill',
            llcrnrlat = -90, llcrnrlon = -180,
            urcrnrlat = 90, urcrnrlon = 180,
           ) # 成圖設定resolution 

#創建邊線
m.drawcoastlines()
#畫出國家，並使用線寬為2 的線條生成分界線。
m.drawcountries(linewidth=2)
#畫出州界
m.drawstates(color='b')
#畫出城市
m.drawcounties(color='darkred')

plt.title('Basemap Tutorial')
plt.show()

```
